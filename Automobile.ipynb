### 1.1 loaded the data from three data files into pandas Dataframes and checked the data is equivalent to raw data files

```python
pd.set_option('display.max_row', 300)
pd.set_option('display.max_columns', 30)
d1 = pd.read_csv("data1.csv")
d2 = pd.read_csv("data2.csv")
d3 = pd.read_csv("data3.csv")
```

### 1.2   Merged the obtained three dataframes into single one that contains all cars, where each car has unique ID and described by 27 attributes 

```python
mergeddata = pd.merge(d1, d2, on = 'id')
final = pd.concat([mergeddata, d3])
final.to_csv("final.csv", index = False)
pd.read_csv("final.csv")
```

### Data Exploration

```python
#counted the number of makes, shows intake of toyata is more 
ax = final['make'].value_counts().plot(kind='bar', figsize=(14,8), title="count of make of car", color='magenta')
ax.set_xlabel("make", fontsize=15, color='magenta')
ax.set_ylabel("count", fontsize=15, color='magenta')   
plt.show()                   
```

```python
# counted engine-type and fount that ohc has more count of engine-type
#ax = final['engine-type'].value_counts().plot(kind='bar', figsize=(14,8), title="count of engine type", color='cyan')
#ax.set_xlabel("engine-type", fontsize=15, color='cyan')
#ax.set_ylabel("count", fontsize=15, color='cyan')   
make_count = final['engine-type'].value_counts().plot(kind='pie')
make_count
plt.axis('equal')
plt.show()
```

```python
#counted body style
ax = final['body-style'].value_counts().plot(kind='barh', figsize=(14,8), title="count of body-style", color='hotpink')
ax.set_xlabel("body-style", fontsize=15, color='hotpink')
ax.set_ylabel("count", fontsize=15, color='hotpink')   
plt.show()
```

```python
#done corelation for all the columns
corr = final.corr()
corr.style.background_gradient(cmap='magma').set_precision(2)
```

```python
#as engine size and price shows the positive value near to 1,so the regression is done between them
#shows its positively co-related, as engine-size increases price increases
plt.figure(figsize=(14,8))
sns.regplot(x="engine-size", y="price", data=final, color='green')
plt.show()
```

```python
#as curb weight and price shows the positive value near to 1,so the regression is done between them
#shows its positively co-related, as curb-weight increases price increases
plt.figure(figsize=(14,8))
sns.regplot(x="curb-weight", y="price", data=final, color='lime')
plt.show()
```

```python
#as highway mileage and price shows the negative value near to -1,so the regression is done between them
#shows its positively co-related, as engine-size increases price increases
plt.figure(figsize=(14,8))
sns.regplot(x="highway-mpg", y="price", data=final, color='crimson')
plt.show()
```

```python
# plotted line graph between engine type and curb-weight, it shows that dohcv has more curb-weight
# as curb weight increases price increases,so the car with dohcv engine-type will be priced more then other cars
et_cw = final[['engine-type','curb-weight']]
et_cw.groupby("engine-type").mean().sort_values("curb-weight", ascending=True).plot.line(figsize=(14,8), color='olive', title='The average curb-weight for each engine type', fontsize = 14)
plt.xlabel('engine-type', fontsize=15, color='olive')
plt.ylabel('curb-weight', fontsize=15, color='olive')
plt.show()
```

```python
#plotted bar graph between make and symboling 
#it can be said that intake of cars for toyata is more as it has average risk rating and less price 
final.groupby("make").symboling.mean().sort_values(ascending=False).plot.bar(figsize=(14,8), color='purple', title='The average purchase price for each make', fontsize = 14)
plt.xlabel('Make', fontsize=15, color='purple')
plt.ylabel('average symboling for insurance risk', fontsize=15, color='purple')
plt.show()
```

```python
#plotted bar graph between make and symboling 
#it can be said that intake of cars for toyata is more as it has average risk rating and less price 
final.groupby("make").price.mean().sort_values(ascending=False).plot.bar(figsize=(14,8), color='red', title='The average purchase price for each make', fontsize = 14)
plt.xlabel('Make', fontsize=15, color='red')
plt.ylabel('Average Price', fontsize=15, color='red')
plt.show()
```

```python
# plotted bar graph between curb-weight and body-style and shown hardtop has more curb-weight
et = final[['body-style','engine-size']]
et.groupby("body-style").mean().sort_values("engine-size", ascending=True).plot.bar(figsize=(12,8), color='blue', title='The average curb-weight for each engine type', fontsize = 14)
plt.xlabel('body-style', fontsize=15, color='blue')
plt.ylabel('engine-size', fontsize=15, color='blue')
plt.show()
```

```python
#as curb weight increases mileage decreases
plt.figure(figsize=(14,8))
sns.regplot(x="highway-mpg", y="curb-weight", data=final, color='crimson')
plt.show()
```

```python
# plotted bar graph between price and body-style and shown hardtop has more price 
# as hardtop as more curb-weight it shows that as curbweight increases price increases
#final.groupby("body-style").price.mean().sort_values(ascending=False).plot.bar(figsize=(14,8), color='teal', title='The average purchase price for each make', fontsize = 14)
ax = sns.boxplot(x="body-style", y="price", data=final)
plt.xlabel('body-style', fontsize=15, color='teal')
plt.ylabel('Price', fontsize=15, color='teal')
plt.show()
```
